```python
# Import the packages we will use in this notebook.
import wrds

import pandas as pd
```

```python
# Replace the username below with your own WRDS username.
username = "junyanxiong"

# Create a WRDS connection.
db = wrds.Connection(wrds_username=username)
```

```python
import numpy as np

ticker = "NKE"
start_date = "2015-01-01"
end_date = "2018-12-31"

sql_nike = f"""
SELECT 
    a.date, 
    a.prc
FROM crsp.dsf a
INNER JOIN crsp.msfhdr b
    ON a.permno = b.permno
WHERE b.htsymbol = '{ticker}'
AND a.date >= '{start_date}'
AND a.date <= '{end_date}'
ORDER BY a.date;
"""

nike = db.raw_sql(sql_nike)
```

```python
import numpy as np

ticker = "UA"

sql_ua = f"""
SELECT 
    a.date, 
    a.prc
FROM crsp.dsf a
INNER JOIN crsp.msfhdr b
    ON a.permno = b.permno
WHERE b.htsymbol = '{ticker}'
AND a.date >= '{start_date}'
AND a.date <= '{end_date}'
ORDER BY a.date;
"""

ua = db.raw_sql(sql_ua)
```

```python
nike = nike.rename(columns={'prc': 'Nike_Price'})
ua = ua.rename(columns={'prc': 'UA_Price'})
```

```python
combined = pd.merge(
    nike[['date', 'Nike_Price']],
    ua[['date', 'UA_Price']],
    on='date',
    how='inner'
)
```

```python
combined['Nike_Return'] = combined['Nike_Price'].pct_change()
combined['UA_Return'] = combined['UA_Price'].pct_change()
```

```python
print("===== Descriptive Statistics (2015 - 2018) =====")
print("\nNike (NKE)")
print(f"Mean Price:    {combined['Nike_Price'].mean():.2f} USD")
print(f"Max Price:     {combined['Nike_Price'].max():.2f} USD")
print(f"Min Price:     {combined['Nike_Price'].min():.2f} USD")
print(f"Avg Daily Return:  {combined['Nike_Return'].mean()*100:.4f} %")
print(f"Volatility (Std):  {combined['Nike_Return'].std()*100:.4f} %")

print("\nUnder Armour (UAA)")
print(f"Mean Price:    {combined['UA_Price'].mean():.2f} USD")
print(f"Max Price:     {combined['UA_Price'].max():.2f} USD")
print(f"Min Price:     {combined['UA_Price'].min():.2f} USD")
print(f"Avg Daily Return:  {combined['UA_Return'].mean()*100:.4f} %")
print(f"Volatility (Std):  {combined['UA_Return'].std()*100:.4f} %")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate cumulative returns
combined['cumulative_nike'] = (1 + combined['Nike_Return']).cumprod()
combined['cumulative_ua'] = (1 + combined['UA_Return']).cumprod()

# Correlation heatmap
corr = combined[['Nike_Return', 'UA_Return']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='RdBu_r', vmin=-1, vmax=1)
plt.title('Return Correlation: Nike vs Under Armour')
plt.show()

# Cumulative returns plot
plt.figure(figsize=(12, 5))
plt.plot(combined['cumulative_nike'], label='Nike', color='#1A9FFF')
plt.plot(combined['cumulative_ua'], label='Under Armour', color='#FF4444')
plt.title('Cumulative Returns Comparison')
plt.legend()
plt.grid(True)
plt.show()

print("Correlation Matrix:")
print(corr)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ----------------------
# Calculate risk/performance metrics
# ----------------------
return_nike = combined['Nike_Return']
return_ua = combined['UA_Return']
trading_days = 252

# Annualized Volatility
nike_vol = return_nike.std() * np.sqrt(trading_days)
ua_vol = return_ua.std() * np.sqrt(trading_days)

# Sharpe Ratio (risk-free rate = 0)
nike_sharpe = return_nike.mean() * trading_days / nike_vol
ua_sharpe = return_ua.mean() * trading_days / ua_vol

# Max Drawdown function
def max_drawdown(return_series):
    cum_ret = (1 + return_series).cumprod()
    peak = cum_ret.cummax()
    drawdown = (cum_ret - peak) / peak
    return drawdown, drawdown.min()

nike_dd, nike_mdd = max_drawdown(return_nike)
ua_dd, ua_mdd = max_drawdown(return_ua)

# ----------------------
# 1. Bar chart comparison of key metrics
# ----------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Annualized Volatility
axes[0].bar(['Nike', 'Under Armour'], [nike_vol, ua_vol], color=['#1A9FFF', '#FF4444'])
axes[0].set_title('Annualized Volatility')
axes[0].set_ylabel('Volatility')
axes[0].grid(axis='y', alpha=0.3)

# Sharpe Ratio
axes[1].bar(['Nike', 'Under Armour'], [nike_sharpe, ua_sharpe], color=['#1A9FFF', '#FF4444'])
axes[1].set_title('Sharpe Ratio (Risk-Free Rate = 0)')
axes[1].set_ylabel('Sharpe Ratio')
axes[1].grid(axis='y', alpha=0.3)

# Max Drawdown
axes[2].bar(['Nike', 'Under Armour'], [nike_mdd, ua_mdd], color=['#1A9FFF', '#FF4444'])
axes[2].set_title('Max Drawdown')
axes[2].set_ylabel('Drawdown')
axes[2].grid(axis='y', alpha=0.3)

plt.suptitle('Nike vs Under Armour Risk & Performance Metrics (2015-2018)', fontsize=14)
plt.tight_layout()
plt.show()

# ----------------------
# 2. Drawdown process over time
# ----------------------
plt.figure(figsize=(12, 5))
plt.plot(combined['date'], nike_dd, label='Nike Drawdown', color='#1A9FFF')
plt.plot(combined['date'], ua_dd, label='Under Armour Drawdown', color='#FF4444')
plt.title('Drawdown Process Over Time')
plt.ylabel('Drawdown')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Make sure date is datetime
combined['date'] = pd.to_datetime(combined['date'])
combined = combined.sort_values('date').reset_index(drop=True)

# Define NBA Finals periods (Jun)
finals_dates = [
    ('2015-06-01', '2015-06-30'),
    ('2016-06-01', '2016-06-30'),
    ('2017-06-01', '2017-06-30'),
    ('2018-06-01', '2018-06-30')
]

# Label each row if it's during finals
combined['during_finals'] = 0
for start, end in finals_dates:
    mask = (combined['date'] >= start) & (combined['date'] <= end)
    combined.loc[mask, 'during_finals'] = 1
```

```python
finals_periods = []

for start, end in finals_dates:
    year = start[:4]
    temp = combined[(combined['date'] >= start) & (combined['date'] <= end)].copy()
    
    if len(temp) < 5:
        continue
    
    # Cumulative return during finals
    cum_nike = (1 + temp['Nike_Return'].dropna()).prod() - 1
    cum_ua = (1 + temp['UA_Return'].dropna()).prod() - 1
    
    # Avg daily return
    avg_nike = temp['Nike_Return'].mean()
    avg_ua = temp['UA_Return'].mean()
    
    # Volatility
    vol_nike = temp['Nike_Return'].std()
    vol_ua = temp['UA_Return'].std()
    
    finals_periods.append({
        'year': year,
        'nike_cum_return': cum_nike,
        'ua_cum_return': cum_ua,
        'nike_daily_avg': avg_nike,
        'ua_daily_avg': avg_ua,
        'nike_vol': vol_nike,
        'ua_vol': vol_ua
    })

finals_df = pd.DataFrame(finals_periods)
print("===== NBA Finals 2015-2018 Stock Performance =====")
print(finals_df.round(4))
```

```python
plt.figure(figsize=(12,6))

for start, end in finals_dates:
    year = start[:4]
    temp = combined[(combined['date'] >= start) & (combined['date'] <= end)].copy()
    temp = temp.dropna(subset=['Nike_Return', 'UA_Return'])
    
    if len(temp) < 3:
        continue
    
    temp['cum_nike'] = (1 + temp['Nike_Return']).cumprod()
    temp['cum_ua'] = (1 + temp['UA_Return']).cumprod()
    
    plt.plot(temp['date'], temp['cum_nike'], label=f'Nike {year}', linewidth=2)
    plt.plot(temp['date'], temp['cum_ua'], label=f'UA {year}', linestyle='--', linewidth=2)

plt.title('Nike vs Under Armour During NBA Finals 2015-2018 (LeBron vs Curry Effect)')
plt.ylabel('Cumulative Return Index')
plt.grid(True)
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
